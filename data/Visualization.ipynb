# Visualization

```python
import datetime
import matplotlib.pyplot as plt

from handling_nan import get_stocks_data
```

```python
# Load data
tickers = ['EFX', 'FB', 'AC.TO', 'MDF.TO', 'AAPL', 'MSFT']
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(3650)
df = get_stocks_data(tickers, start_date, end_date, handle_nan='backfill')
```

```python
df.plot()
```

```python
df_standardized = df / df.min()
```

```python
df_standardized.plot()
```

```python
df_standardized.plot(subplots=True, layout=(3,2))
```

```python
# Matplotlib Plotting
daily_return = df.pct_change()
fig, ax = plt.subplots()
plt.style.use('seaborn-dark')
ax.set(title='Daily return on tech stocks', xlabel="Tech Stocks")
plt.bar(daily_return.columns, daily_return.mean())
```
